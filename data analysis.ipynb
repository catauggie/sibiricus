```python
import pandas as pd
import matplotlib.pyplot as plt

chars = ['SoilTemp', 'Tair', 'Evap', 'Rainf', 'SoilMoi']
labels_area = ['Infected Area', 'Uninfected Area']
labels_unit = ['$^\circ$C', ' $^\circ$C', r'$\frac{mm}{month}$', 'mm', '%']
labels_chars = ['Soil Temperature', 'Air Temperature', 'Evaporation', 'Rainfall', 'Soil Moisture']
years_outbreaks = list(df['Год вспышки'].unique())

for a in list(range(len(labels_area))):
    for c in range(len(chars)):
        char_columns = [col for col in df.columns if chars[c] in col]

        for y in range(len(years_outbreaks)):
            df_spec_char = df[(df['Вспышка Да/Нет Эталон'] == a) & (df['Год вспышки']==years_outbreaks[y])][char_columns]
            # Assuming you have a dictionary with the data
            data_spec = df_spec_char.mean()

            # Create DataFrame
            df_spec = pd.DataFrame(data_spec.items(), columns=['Month', 'Value'])

            # Extract month number from the column names and convert them to numeric for sorting
            df_spec['Month'] = df_spec['Month'].str.extract(r'_(\d+)_').astype(int)

            # Group by month and aggregate values into lists
            grouped = df_spec.groupby('Month')['Value'].apply(list)

            # Convert the grouped series to a dictionary
            grouped_dict = grouped.to_dict()

            grouped_df = pd.DataFrame(grouped_dict)
            
            # Applying conditions
            if chars[c] in ['SoilTemp', 'Tair']:
                grouped_df = grouped_df - 273.15
            elif chars[c] in ['Evap', 'Rainf']:
                grouped_df = grouped_df * 100000
            elif chars[c] == 'SoilMoi':
                grouped_df = grouped_df * 100

            # English month names
            eng_month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
            years_before_spec_descending = [years_outbreaks[y] - i - 1 for i in range(7)]
            months = list(grouped_df.columns)
            selected_months = [eng_month_names[m - 1] for m in months]

            # Create plot
            plt.figure(figsize=(10, 6))

            for index, row in grouped_df.iterrows():
                plt.plot(row, label=f'Row {index}')

            plt.title(f'{labels_chars[c]} Variation during {grouped_df.shape[0]} years before outbreak of {years_outbreaks[y]} \n{labels_area[a]}')
            plt.xlabel('Month')
            plt.ylabel(f'{labels_chars[c]}, {labels_unit[c]}')
            plt.legend(years_before_spec_descending)
            plt.grid(True)
            plt.xticks(months, selected_months, rotation=45)
            plt.tight_layout()
            plt.savefig(f'{chars[c]}_{grouped_df.shape[0]}_years_{[a]}_outbreak_{years_outbreaks[y]}.pdf')
            plt.show()
```

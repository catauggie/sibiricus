```python
import folium
import pandas as pd

# Sample data
data = df.loc[0:100][['X', 'Y']]#{
   # 'X': [95.64974245, 96.01209857, 95.64548277, 95.64339089, 95.65116502],
 #   'Y': [55.03903355, 55.04125302, 55.02311747, 55.01771424, 55.01479225]
#}#

# Create DataFrame
df = pd.DataFrame(data)

# Initialize the map
mymap = folium.Map(location=[df['Y'].mean(), df['X'].mean()], zoom_start=10)

# Add points to the map
for index, row in df.iterrows():
    folium.CircleMarker([row['Y'], row['X']], radius=5, color='blue', fill=True, fill_color='blue').add_to(mymap)

# Save the map to an HTML file
mymap.save("map_with_dots.html")

# Display the map
mymap
```

```python
import pandas as pd
df = pd.read_excel('data.xlsx')
```

```python
chars = ['SoilTemp', 'Tair', 'Evap', 'Rainf', 'SoilMoi']
char_lists = []
for c in range(len(chars)):
    char_columns = [col for col in df.columns if chars[c] in col]
    char_lists.append(char_columns)
```

```python
char_lists
```

```python
import re
import calendar

data = ['Tair_f_tavg_5_май',
        'Tair_f_tavg_6_июнь',
        'Tair_f_tavg_7_июль',
        'Tair_f_tavg_8_август',
        'Tair_f_tavg_5_май.1',
        'Tair_f_tavg_6_июнь.1',
        'Tair_f_tavg_7_июль.1',
        'Tair_f_tavg_8_август.1',
        'Tair_f_tavg_5_май.2',
        'Tair_f_tavg_6_июнь.2',
        'Tair_f_tavg_7_июль.2',
        'Tair_f_tavg_8_август.2']

# Создаем словарь, где ключами будут названия месяцев на русском языке, а значениями будут списки элементов для каждого месяца
grouped_data = {}

for item in data:
    # Извлекаем номер месяца с помощью регулярного выражения
    month_number_match = re.search(r'\d+', item)
    if month_number_match:
        month_number = int(month_number_match.group())
        month_name = calendar.month_name[month_number].lower()  # Получаем название месяца на русском языке
        if month_name in grouped_data:
            grouped_data[month_name].append(item)
        else:
            grouped_data[month_name] = [item]

# Преобразуем словарь в список списков
result = list(grouped_data.values())

print(result)

```

```python
# Создаем список для хранения результатов для каждого списка в char_lists
result_lists = []

for data in char_lists:
    grouped_data = {}

    for item in data:
        # Извлекаем номер месяца из строки
        month_number_match = re.search(r'_(\d+)_', item)
        if month_number_match:
            month_number = int(month_number_match.group(1))
            month_name = calendar.month_name[month_number].lower()  # Получаем название месяца на русском языке
            if month_name in grouped_data:
                grouped_data[month_name].append(item)
            else:
                grouped_data[month_name] = [item]

    # Преобразуем словарь в список списков и добавляем его в результаты
    result_lists.append(list(grouped_data.values()))
```

```python
result_lists[0][0]
```

```python
averages_df = pd.DataFrame()

# Функция для вычисления среднего значения по группе элементов
def compute_group_average(group, df):
    # Создаем временный DataFrame, чтобы вычислить среднее значение по строкам
    temp_df = pd.DataFrame()
    for item in group:
        # Добавляем столбец с данными из item
        temp_df[item] = df[item]
    # Вычисляем среднее значение по строкам для каждого элемента в группе
    return temp_df.mean(axis=1)

# Перебираем группы во всех уровнях вложенности result_lists
for i, outer_group in enumerate(result_lists):
    for j, inner_group in enumerate(outer_group):
        # Вычисляем среднее значение для текущей группы
        group_average = compute_group_average(inner_group, df)
        # Добавляем средние значения в новый столбец в общий DataFrame averages_df
        averages_df[f'average_group_{i+1}_{j+1}'] = group_average
```

```python
import re

# Предположим, что у нас есть список списков result_lists

# Создадим пустой список для хранения результатов
extracted_data = []

# Перебираем элементы result_lists
for sublist in result_lists:
    # Перебираем элементы внутри каждого sublist
    for item in sublist:
        # Извлекаем первое слово из item
        first_word = item[0].split('_')[0]
        # Извлекаем цифру из item
        number = re.search(r'_(\d+)_', item[0])
        if number:
            number = number.group(1)
            # Добавляем первое слово и цифру в список extracted_data
            extracted_data.append(first_word + '_' + number)

print(extracted_data)
```

```python
# Добавляем названия колонок из extracted_data
averages_df.columns = extracted_data
```

```python
# Предположим, что у вас есть DataFrame df с колонками "Тип леса" и "Влажность почвы"

# Применяем OHE к столбцу "Тип леса" и сохраняем результат в новый DataFrame
df_tipohe = pd.get_dummies(df['Тип леса'])

# Применяем OHE к столбцу "Влажность почвы" и сохраняем результат в новый DataFrame
df_humidityohe = pd.get_dummies(df['Влажность почвы'])

# Объединяем новые DataFrame вместе
new_df_ohe = pd.concat([df_tipohe, df_humidityohe], axis=1)
```
